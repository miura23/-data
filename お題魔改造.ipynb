```python
import requests
import pandas as pd
import csv
import json

def get_data(limit, offset):
    url = "https://api.p2pquake.net/v2/jma/quake"
    params = {"limit": limit, "offset": offset}
    response = requests.get(url, params=params)
    data_list = response.json()
    return data_list

limit = 100
data = []
for i in range(3): #(1)をすれば100件
    offset = i * limit
    data += get_data(limit=limit, offset=offset)
```

```python
# forを使ったやり方
num = 0
for data in data:
    num += 1
    print(num, data["created_at"],
          data["earthquake"]["hypocenter"]["magnitude"],
          data["earthquake"]["hypocenter"]["name"])
```

```python
#pandas使用
#lambda式は、関数に名前を付けずに定義できる
df = pd.DataFrame(data)
df["created_at"] = df["created_at"]
df["magnitude"] = df["earthquake"].apply(lambda x: x["hypocenter"]["magnitude"])
df["name"] = df["earthquake"].apply(lambda x: x["hypocenter"]["name"])
df["points_count"] = df["points"].apply(lambda x: len(x))
print(df[["created_at", "magnitude", "name","points_count"]])
```

```python
#震度3以上を表示
df = pd.DataFrame(data)
df["created_at"] = df["created_at"]
df["magnitude"] = df["earthquake"].apply(lambda x: x["hypocenter"]["magnitude"])
df["name"] = df["earthquake"].apply(lambda x: x["hypocenter"]["name"])
df["scale"] = df["points"].apply(lambda x: x[0]["scale"] if x else None)
df["points_count"] = df["points"].apply(lambda x: len(x))
df2 = df[df["scale"] >= 30]
print(df2[["created_at", "magnitude", "name","scale","points_count"]])
```

```python
import folium
#緯度、経度dfに追加
df['latitude'] = df['earthquake'].apply(lambda x: x['hypocenter']['latitude'])
df['longitude'] = df['earthquake'].apply(lambda x: x['hypocenter']['longitude'])
#緯度、経度をもとにマッピングする
chizu = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=5)

for lat, lon in zip(df['latitude'], df['longitude']):
    folium.Marker([lat, lon]).add_to(chizu)
```
